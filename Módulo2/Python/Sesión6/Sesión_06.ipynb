# Creando dataset

```
import pandas as pd
import requests
import time
```

```
endpoint = "https://api.nasa.gov/neo/rest/v1/neo/browse"
payload = {"api_key": "3TORfqwWIrHnAaqCOTeb8H9Y8jEUmGpCZ049xucd"}
```

```
dict_datos = {}

for i in range(0,10):

  try:

    time.sleep(5) #Se pone el tiempo de espera para evitar el bloqueo por exceso de requesr
    r = requests.get(endpoint, params = payload, timeout = 5)

    if r.status_code == 200:

      json = r.json()
      data = json["near_earth_objects"]
      dict_datos[i] = data

      new_link = json["links"]["next"]
      endpoint = new_link

  except:
    
    continue
```

```
for key in dict_datos:

  normalized = pd.json_normalize(dict_datos[key])
  dict_datos[key] = normalized
```

```
dict_datos
```

```
lista_de_dataframes = []

for key in dict_datos:

  lista_de_dataframes.append(dict_datos[key])
```

```
df_completo = pd.concat(lista_de_dataframes, axis = 0). reset_index(drop = True)
```
