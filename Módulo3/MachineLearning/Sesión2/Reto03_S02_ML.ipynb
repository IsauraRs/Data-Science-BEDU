# Leave-one-out Cross validation

```python
import pandas as pd
from sklearn.model_selection import KFold
```

```python
data = pd.read_csv("/content/drive/MyDrive/bitstampUSD_1-min_data_2012-01-01_to_2020-12-31.csv" , sep = ',')
```

```python
data = data.apply(pd.to_numeric, errors = 'coerce')
data = data.dropna()
```

```python
data = data.reset_index(drop = True)
```

```python
x = data.drop(columns= ['Timestamp', 'Close'])
y = data.Close
```

```python
data = data

K = data.shape[0] # K = numero de muestras, toma todo el conjunto
aleatorio = True

kfold = KFold(K, aleatorio)
ciclo = 1
for indices_train, indices_test in kfold.split(data):
    print("Ciclo: "+str(ciclo))
    print("\t datos para entrenamiento:"+str(data.iloc[indices_train]))
    print("\t datos para prueba:"+str(data.iloc[indices_test]))
    ciclo+=1
```
